```python
for i in range(11,122):    
    urlpage='http://www.dhcc.ae/Portal/en/healthcare/doctors.aspx#page='+str(i)
    driver.get(urlpage)
    time.sleep(10)
    soup = BeautifulSoup(driver.page_source, 'lxml')
    f0 =soup.find_all('li', class_="fi-pageme-page current")
    f = f0[0].find_all('tr', class_="data")
    for q in range(0,len(f)):
        data=['null']*6
        clinic0=f[q].find('td', class_="second")
        clinic=(clinic0.text).encode('ascii', 'ignore')
        if len(clinic)>0:
            Name0=f[q].find('td', class_="first")
            Name=(Name0.text).encode('ascii', 'ignore')
            #print(Name)
            data[0]=Name.replace(","," &")
            #print(clinic)
            data[1]=clinic.replace(","," &")
            specialty0=f[q].find('td', class_="third")
            specialty=specialty0.text
            data[2]=specialty.replace(","," &")
            #print(specialty)
            language0=f[q].find('td', class_="fourth")
            language=(language0.text).encode('ascii', 'ignore')
            #print(language)
            data[3]=language.replace(","," &")
            gender0=f[q].find('td', class_="fifth")
            gender=gender0.text
            #print(gender)
            data[4]=gender
            Id_link=(Name0.find('a', href=True)).attrs['href']
            #print(Id_link)
            data[5]=("http://www.dhcc.ae/Portal/"+Id_link).encode('ascii', 'ignore')
            print("+++++++++++++++")
            print(data)
            afaq=[data]
            csvfile = open('output2.csv', 'a')
            for row in afaq:
                #line = ','.join(row)
                values = ','.join(str(v) for v in row)
                csvfile.write(values + '\n')
```
