```python
for z in range(1,319):
    url = 'https://www.rajamobil.com/jual/mobil/bekas?page='+str(z)
    res = requests.get(url,headers={"User-Agent":"Mozilla/5.0"})
    soup = BeautifulSoup(res.text,"lxml")
    searchcars=soup.find_all('div',class_="list-mobil")

    for i in range(0,len(searchcars)):
        datastr=['-']*6
        brand=searchcars[i].find('h2',class_="firstTitle").text
        datastr[0]=brand

        Model=searchcars[i].find('h3',class_="secondTitle").text
        datastr[1]=Model

        spek=searchcars[i].find('ul',class_="umum-spek")
        all_li=spek.find_all('li')

        cccar=all_li[1].text
        datastr[2]=cccar

        trcar=all_li[2].text
        datastr[3]=trcar

        mlcar=all_li[3].text
        datastr[4]=mlcar

        price=searchcars[i].find('div',class_="thirdTitle font_leasing_text").text
        datastr[5]=price
        print datastr,len(savedata1)
        savedata1.append(datastr)
```
