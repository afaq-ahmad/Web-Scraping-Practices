```python
import string
string.ascii_lowercase
alpha=list(string.ascii_lowercase)
```

```python
for w in range(2,11):
    for q in range(0,len(alpha)):
        for k in range(0,len(alpha)):
            print w,q,k        
            urlpage='http://www.info718.com/residential/search?q='+alpha[w]+alpha[q]+alpha[k]
    #         print urlpage
            sleep(2)
            headers = {'User-Agent':'Mozilla/5.0'}
            page = requests.get(urlpage, headers=headers, verify=False)
            data = page.content
            soup = BeautifulSoup(data, "html.parser")

            business=soup.find_all('div',class_="business")
            print (len(business))

            for i in range(2,len(business)):
                data0=['-']*5
                name0=business[i].find('span',class_="name")
                if name0 is not None:
                    name=name0.text
                    data0[0]=name
                    Last_name=name.split()[-1]
                    data0[1]=Last_name
                    First_name=' '.join(name.split()[:-1])
                    data0[2]=First_name


                    phone0=business[i].find('span',class_="phone")
                    if phone0 is not None:
                        phone=phone0.text
                        data0[3]=phone

                        address=business[i].find('span',class_="address").text
                        data0[4]=address
                        data1.append(data0)
                        csvWriter = csv.writer(open('output_3.csv', 'a'))
                        csvWriter.writerow(data0)
            print   (len(data1))
```
