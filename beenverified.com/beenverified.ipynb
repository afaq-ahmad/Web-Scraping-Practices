```python
from openpyxl import load_workbook

from bs4 import BeautifulSoup
from selenium import webdriver
from time import sleep
import csv
from random import randint
import json, io

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import NoAlertPresentException
from selenium.webdriver.common.action_chains import ActionChains

import urllib
import urllib3
import requests
import json, io
from bs4 import BeautifulSoup
urllib3.disable_warnings()
header = {'User-Agent':'Mozilla/5.0'}


chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--user-agent="Mozilla/5.0')
chrome_options.add_argument("user-data-dir=selenium")
driver = webdriver.Chrome(chrome_options=chrome_options, executable_path=r'/Users/Name/Downloads/Compressed/chromedrives/chromedriver.exe')


cookies = json.load(open('cookiesdict.txt'))
for cookie in cookies:
    driver.add_cookie(cookie)
    
    
driver.get('https://www.beenverified.com/f/login')
```

```python
list0=[]
wb = load_workbook(filename='new_big_file2.xlsx', read_only=True)
ws = wb['Sheet']
for row in ws.rows:
    list0.append((row))
# del list0[0]

bigset=[]
for i in range(0,len(list0)):
    set1=[]
    parcelid=list0[i][0].value
    set1.append(parcelid)
    owner10=list0[i][2].value
    owner1=owner10.split(',')
    firstname=owner1[-1]
    set1.append(firstname)
    lastname=owner1[0]
    set1.append(lastname)
    bigset.append(set1)
    if len(list0[i])==4:
        set2=[]
        set2.append(parcelid)
        owner20=list0[i][3].value
        owner2=owner20.split(',')
        firstname=owner2[-1]
        set2.append(firstname)
        lastname=owner2[0]
        set2.append(lastname)
        bigset.append(set2)
```

```python
savedata1=[]
def userdetails():
    savedata0=[]
    fullname=driver.find_element_by_xpath('//*[@id="overview"]/div/div[1]/div[2]/div[1]/div[1]/div[2]').text
    savedata0.append(fullname)
    Born=driver.find_element_by_xpath('//*[@id="overview"]/div/div[1]/div[2]/div[2]/div[1]/div[2]').text
    savedata0.append(Born)
    CurrentAddress=driver.find_element_by_xpath('//*[@id="overview"]/div/div[1]/div[2]/div[2]/div[2]/div[2]').text
    savedata0.append(CurrentAddress)
    
    so1=BeautifulSoup(driver.page_source, 'lxml')
    phonecollector=[]
    emailcollector=[]

    phone0=so1.find('div',id="phone-numbers")
    len(phone0)
    phone=phone0.find_all('h4')
    for m in range(1,len(phone)-1):
        phonecollector.append(phone[m].text)
    
    email0=so1.find('div',id="email-addresses")
    len(email0)
    email=email0.find_all('h4')
    for n in range(1,len(email)-1):
        emailcollector.append(email[n].text)
    savedata1.append(savedata0+phonecollector+emailcollector)
    print savedata1
```
