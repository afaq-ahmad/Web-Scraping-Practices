```python
for u in sitelinks:
    
    driver.get(u)
    sleep(7)
    if len(driver.current_url)>56:
        try:
            while True:
                so1=BeautifulSoup(driver.page_source, 'lxml')
                business=so1.find_all('div',id="geo-business")
                business_info=business[0].find_all(class_="business-attribution__business-info")
                previous_length=len(business_info)


                for l in range(3):
                    driver.execute_script("window.scrollTo(0, 1000000)")
                    sleep(3)
                so1=BeautifulSoup(driver.page_source, 'lxml')
                business=so1.find_all('div',id="geo-business")
                business_info=business[0].find_all(class_="business-attribution__business-info")
                print len(business_info)
                if previous_length==len(business_info):
                    break


            for l in range(0,len(business_info)):
                datatemp=[]

                linkc=business_info[l].find('a').get('href')
                datatemp.append(linkc)
                name=business_info[l].find('strong').text

                datatemp.append(name)

                address=business_info[l].find_all('div')[2].text.replace('\n','')
                datatemp.append(address)
                try:
                    category=business_info[l].find_all('div')[3].text.replace('\n','')
                    datatemp.append(category)
                except:
                    print 'l'
                print datatemp
                fulldatastr.append(datatemp)
        except:
            pass
```
