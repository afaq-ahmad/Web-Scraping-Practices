```python
list0=[]
wb = load_workbook(filename='to_complete.xlsx', read_only=True)
ws = wb['Sheet1']
for row in ws.rows:
    list00=[]
    for kl in range(len(row)):
        
        list00.append(row[kl].value)
    list0.append(list00)
```

```python
savedata1=[]
```

```python
for i in range(2,len(list0)):

    inpu=driver.find_elements_by_name('s')[0]
    inpu.clear()
    sleep(1)
    inpu.send_keys(list0[i][0])
    sleep(1)
    driver.find_element_by_class_name('input-group-button').click()
    sleep(3)
    
    so1=BeautifulSoup(driver.page_source, 'lxml')

    datastr=['-']*6

    acronym=so1.find_all('strong',class_="acronym-key")
    for p in range(len(acronym)):
        if acronym[p].text=='IATA Code':
            IATA=acronym[p].next_element.next_element.text
            datastr[0]=IATA

        if acronym[p].text=='ICAO Code':
            ICAO=acronym[p].next_element.next_element.text
            datastr[1]=ICAO


    Airport_name=so1.find_all('h1',class_="airport-title")[0].next_element.replace('\n','').replace('  ','').replace(u'\xa0', u' ')
    datastr[2]=Airport_name

    details_city=so1.find(class_="header clearfix").find_all('p',class_="subheader")[0].text.replace('\n','').replace('  ','')
    datastr[3]=details_city

    latitude=so1.find('input', id="latitude").get('value')
    datastr[4]=latitude
    longitude=so1.find('input', id="longitude").get('value')
    datastr[5]=longitude
    print datastr
    savedata1.append(datastr)
    driver.back()
    sleep(3)
```

```python
import warnings
from openpyxl import Workbook

wb = Workbook(write_only=True)
ws = wb.create_sheet()

# now we'll fill it with 100 rows x 200 columns
for irow in savedata1:
    ws.append(irow)
# save the file
wb.save('results.xlsx')
```
