```python
from os.path import basename
import os
from urlparse import urlsplit
import urllib2


#dir is not keyword
def makemydir(whatever):
    try:
        os.makedirs(whatever)
    except OSError:
        pass
    # let exception propagate if we just can't
    # cd into the specified directory
    os.chdir(whatever)
    
```

```python
for z in range(0,len(linksto_profiles)):
    
    mainpage=requests.get(linksto_profiles[z])
    so0=bs4.BeautifulSoup(mainpage.text,'lxml')
    

    datastr=['-']*12
    car_name=so0.find('h1',itemprop="name").text
    datastr[0]=car_name
    

    Manufactering_year=so0.find_all('tbody')[0].find_all('p',itemprop="dateVehicleFirstRegistered")[0].text
    datastr[1]=Manufactering_year


    Meter_reading=so0.find_all('tbody')[0].find_all('p',itemprop="mileageFromOdometer")[0].text
    datastr[2]=Meter_reading

    vehicle_Transmission=so0.find_all('tbody')[0].find_all('p',itemprop="vehicleTransmission")[0].text
    datastr[3]=vehicle_Transmission

    car_features=so0.find_all('ul',class_="list-unstyled ul-featured clearfix")[0]

    car_featuresli=car_features.find_all('li')

    for r in range(len(car_featuresli)):
        if car_featuresli[r].text=='Registered City':
            Registered_City=car_featuresli[r+1].text
            datastr[4]=Registered_City
        if car_featuresli[r].text=='Color':
            Color=car_featuresli[r+1].text
            datastr[5]=Color
        if car_featuresli[r].text=='Assembly':
            Assembly=car_featuresli[r+1].text
            datastr[6]=Assembly
        if car_featuresli[r].text=='Engine Capacity':
            Engine_Capacity=car_featuresli[r+1].text
            datastr[7]=Engine_Capacity
        if car_featuresli[r].text=='Body Type':
            Body_Type=car_featuresli[r+1].text
            datastr[8]=Body_Type
        if car_featuresli[r].text=='Ad Ref #':
            Ad_Ref=car_featuresli[r+1].text
            datastr[9]=Ad_Ref
    car_price=so0.find_all('div',class_="price-box generic-blue")[0].text
    datastr[10]=car_price
    datastr[11]=linksto_profiles[z]
    print datastr,z
    datastorage.append(datastr)
```
