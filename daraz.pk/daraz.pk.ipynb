```python
overall_Save=[]
tags=['name','brandName','review','ratingScore','discount','priceShow','originalPrice','price','brandId','itemId','sellerName','location','restrictedAge','sellerId','productUrl','image','voucherId','description']
overall_Save.append(tags)
import glob

Json_files=glob.glob('*.json')


for mj in range(len(Json_files)):


    with open(Json_files[mj]) as f:
        data = json.load(f)

    for k in range (len(data['mods']['listItems'])):

        data_Save=[None]*18

        name=data['mods']['listItems'][k]['name']
        data_Save[0]=name
        brandName=data['mods']['listItems'][k]['brandName']
        data_Save[1]=brandName
        review=data['mods']['listItems'][k]['review']
        data_Save[2]=review
        ratingScore=data['mods']['listItems'][k]['ratingScore']
        data_Save[3]=ratingScore
        try:

            discount=data['mods']['listItems'][k]['discount']
            data_Save[4]=discount
        except:
            print 'no discount'

        priceShow=data['mods']['listItems'][k]['priceShow']
        data_Save[5]=priceShow
        try:
            originalPrice=data['mods']['listItems'][k]['originalPrice']
            data_Save[6]=originalPrice
        except:
            print 'No originalPrice'


        price=data['mods']['listItems'][k]['price']
        data_Save[7]=price
        brandId=data['mods']['listItems'][k]['brandId']
        data_Save[8]=brandId
        itemId=data['mods']['listItems'][k]['itemId']
        data_Save[9]=itemId
        Seller_Name=data['mods']['listItems'][k]['sellerName']
        data_Save[10]=Seller_Name
        try:
            location=data['mods']['listItems'][k]['location']
            data_Save[11]=location
        except:
            print 'no location'


        restrictedAge=data['mods']['listItems'][k]['restrictedAge']
        data_Save[12]=restrictedAge

        sellerId=data['mods']['listItems'][k]['sellerId']
        data_Save[13]=sellerId

        product_url=data['mods']['listItems'][k]['productUrl']
        data_Save[14]=product_url

        image_url=data['mods']['listItems'][k]['image']
        data_Save[15]=image_url

        voucherId=data['mods']['listItems'][k]['voucherId']
        data_Save[16]=voucherId
        data_Save[17]=Json_files[mj]
        
        description=data['mods']['listItems'][k]['description']


        overall_Save.append(data_Save+description)
```

```python
import warnings
from openpyxl import Workbook

wb = Workbook(write_only=True)
ws = wb.create_sheet()

# now we'll fill it with 100 rows x 200 columns
for irow in overall_Save:
    for i in range(len(irow)):
        try:
            irow[i]=irow[i].encode('utf8')
        except:
            pass
    irow[16]=irow[16].replace('x','')
    ws.append(irow)
# save the file
wb.save('mults.xlsx')
```
