```python
for q in range(6692,7000):
    urlpage1=idlink[q]
    page1 = requests.get(urlpage1, headers=headers, verify=False)
    dataforsoup1 = page1.content 
    soup1 = BeautifulSoup(dataforsoup1, "html.parser")
    sleep(3)

    vcard=soup1.find_all('div',class_="item vcard")[0]

    dataset=['-']*10
    dataset[0]=urlpage1
    company_name=vcard.find('h1').text
    dataset[1]=company_name

    street_address=vcard.find('span',class_="street-address").text
    dataset[2]=street_address


    locality=vcard.find('span',class_="locality").text
    dataset[3]=locality

    region=vcard.find('span',class_="region").text
    dataset[4]=region

    postal=vcard.find('span',class_="postal-code").text
    dataset[5]=postal


    tel=vcard.find('span',class_="tel").text
    dataset[6]=tel

    Established0=vcard.find(id="p-est")
    if Established0 is not None:
        Established=Established0.text
        dataset[7]=Established

    web=vcard.find(id="p-web").text
    dataset[8]=web

    email=vcard.find(id="p-email").text
    dataset[9]=email
    datasetmm.append(dataset)
    print (q,len(datasetmm))
    
    spamWriter = csv.writer(open('output5000_10000.csv', 'a'))
    spamWriter.writerow(dataset)
```
