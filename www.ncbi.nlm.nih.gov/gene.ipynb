```python
s = BeautifulSoup(driver.page_source, 'lxml')
result0=s.find_all('tr', class_="rprt")
for i in range(0,len(result0)):
    linkdata=[0]
    if result0[i].find('em').text=='Homo sapiens':
        link0=result0[i].find_all('a',href=True)
        link=link0[0].attrs['href']
        id=((result0[i].find('span',class_="gene-id").text).split())[-1]
        print(link)
        linkdata=link
        spamWriter = csv.writer(open('outputlink.csv', 'a'))
        spamWriter.writerow(linkdata)
        linkdata.append((link,id))
```

```python
linkdata=[]
for p in range(11):
    s = BeautifulSoup(driver.page_source, 'lxml')
    result0=s.find_all('tr', class_="rprt")
    for i in range(0,len(result0)):
        linkdatasave=[0]*2
        if result0[i].find('em').text=='Homo sapiens':
            link0=result0[i].find_all('a',href=True)
            link=link0[0].attrs['href']
            id=((result0[i].find('span',class_="gene-id").text).split())[-1]
            linkdata.append((link,id))
            linkdatasave[0]=id
            linkdatasave[1]=link
            spamWriter = csv.writer(open('outputlink.csv', 'a'))
            spamWriter.writerow(linkdatasave)

    cc=driver.find_elements_by_name('EntrezSystem2.PEntrez.Gene.Gene_ResultsPanel.Entrez_Pager.Page')
    if cc[2].text=='Next >':
        cc[2].click()
        sleep(10)
    print(p)
```
