```python
def datagetter(link1):
    data=[0]*11
    link2='https://turnaround.org'
    suburl=link2+link1
    print(suburl)
    page1 = requests.get(suburl, headers=headers, verify=False)
    data1 = page1.content 
    soup1 = [BeautifulSoup(data1, "html.parser")]

    data[0]=suburl
    prof0=soup1[0].find_all('div', class_="group-member-profile-info")
    name0=prof0[0].find('div', class_="field-name-field-display-name")
    name1=(((name0.text).replace("\t","")).replace("\n",""))
    name1=name1.encode('ascii', 'ignore')
    name1=name1.split()
    firstname=name1[0]
    secondname0=name1[1:len(name1)]
    secondname=' '.join(map(str, secondname0))
    secondname=secondname.replace(",",".")
    data[1]=firstname.encode('ascii', 'ignore')
    data[2]=secondname.encode('ascii', 'ignore')
    print(firstname)
    print(secondname)

    company0=prof0[0].find('div', class_="field-name-user-firm-name")
    if company0 is not None:        
        company=(((company0.text).replace("\t","")).replace("\n",""))
        company=company.replace(",",".")
        data[3]=company.encode('ascii', 'ignore')
        print(company)
    else:
        data[3]='-'

    address0=prof0[0].find('div', class_="field-name-field-person-address")
    #print(address0)
    if address0 is not None:
        
        streetAddress=address0.find('span', itemprop="streetAddress")
        if streetAddress is not None:
            streetAddress=(streetAddress.text).replace(","," ")
            data[4]= streetAddress.encode('ascii', 'ignore')
            print(streetAddress)
        else:
            data[4]='-'

        addressLocality=address0.find('span', itemprop="addressLocality")
        if addressLocality is not None:
            addressLocality=((((addressLocality.text).replace(","," ")).replace("\t","")).replace("\n",""))
            addressLocality=addressLocality.replace("  ","")
            data[5]=addressLocality.encode('ascii', 'ignore')
            print(addressLocality)
        else:
            data[5]='-'

        addressRegion=address0.find('span', itemprop="addressRegion")
        if addressRegion is not None:        
            addressRegion=(addressRegion.text).replace(","," ")
            data[6]=addressRegion.encode('ascii', 'ignore')
            print(addressRegion)
        else:
            data[6]='-'

        postalCode=address0.find('span', itemprop="postalCode")
        if postalCode is not None:        
            postalCode=(postalCode.text).replace(","," ")
            data[7]=postalCode.encode('ascii', 'ignore')
            print(postalCode)
        else:
            data[7]='-'


        addressCountry=address0.find('div', itemprop="addressCountry")
        if addressCountry is not None:
            addressCountry=(addressCountry.text).replace(","," ")
            data[8]=addressCountry.encode('ascii', 'ignore')
            print(addressCountry)
        else:
            data[8]='-'
    else:
        data[4]='-'
        data[5]='-'
        data[6]='-'
        data[7]='-'
        data[8]='-'

    phone0=prof0[0].find('div', class_="field-name-field-number")
    if phone0 is not None:
        phone=(((phone0.text).replace("\t","")).replace("\n",""))
        phone=phone.replace(",",".")
        data[9]=phone
        print(phone)
    else:
        data[9]='-'

    email0=prof0[0].find('div', class_="field-name-user-email-link")
    if email0 is not None:
        email=(((email0.text).replace("\t","")).replace("\n",""))
        print(email)
        email=email.replace(","," ")
        data[10]=email.encode('ascii', 'ignore')
    else:
        data[10]='-'
    
    afaq=[data]
    csvfile = open('outputre.csv', 'a')
    for row in afaq:
        #line = ','.join(row)
        values = ','.join(str(v) for v in row)
        csvfile.write(values + '\n')
```
