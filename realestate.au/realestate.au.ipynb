```python
url='https://www.realestate.com.au/buy/property-townhouse-unit+apartment-in-glenelg%2c+sa+5045/list-1?includeSurrounding=false&source=location-search'
import csv
import requests
from bs4 import BeautifulSoup
import urllib3
urllib3.disable_warnings()
from collections import OrderedDict
import time

houseslinks=[]

headers = {'User-Agent':'Mozilla/5.0'}
page = requests.get(url, headers=headers, verify=False)
data = page.content
soup = BeautifulSoup(data, "html.parser")

while(True):
    linkh0=soup.find_all('a',class_="detailsButton",href=True)
    #print linkh0
    for i in range(len(linkh0)):
        linkh1=linkh0[i].attrs['href']
        linkinitail='https://www.realestate.com.au'
        linkh=linkinitail+linkh1
        print (linkh)
        houseslinks.append(linkh)
    nextlink0=soup.find_all('li',class_="nextLink")
    if len(nextlink0)>0:
        nextlink1=nextlink0[0].find('a',href=True)
        nextlink2=nextlink1.attrs['href']
        nextlink=linkinitail+nextlink2
        page = requests.get(nextlink, headers=headers, verify=False)
        data = page.content
        soup = BeautifulSoup(data, "html.parser")
        
    else:
        break
list1=list(OrderedDict.fromkeys(houseslinks))
print ("Total number of houses found>>>",len(list1))
```

```python
for z in range(0,len(list1)):
    data=['null']*15
    url1=list1[z]
    page = requests.get(url1, headers=headers, verify=False)
    data1s = page.content
    soup = BeautifulSoup(data1s, "html.parser")
    property_id0=soup.find('span',class_="property_id")
    if property_id0 is None:
        linkh0=soup.find_all('a',class_="detailsButton",href=True)
        for i in range(len(linkh0)):
            linkh1=linkh0[i].attrs['href']
            linkinitail='https://www.realestate.com.au'
            linkh=linkinitail+linkh1
            list1.append(linkh)
        print ("This House advertisement has innerr"+str(len(linkh0))+"house mentioned, those scraped at last")
        print (url1)
        
        continue 
    property_id=property_id0.text
    property_id=property_id.split()[-1]
    data[0]=property_id

    propertyType=soup.find('span',class_="propertyType").text
    data[1]=propertyType

    price0=soup.find('p',class_="priceText")
    if price0 is not None:
        price=price0.text
        data[2]=price


    agentName=soup.find_all('p',class_="agentName")
    if agentName is not None:
        agentName=agentName[0].text
        data[3]=agentName

    phone=soup.find_all('li',class_="printHidden phoneHidden")[0].text
    data[4]=phone

    address0=soup.find_all('div',id="listing_address")
    streetAddress=address0[0].find('span',itemprop="streetAddress").text
    data[5]=streetAddress

    addressLocality=address0[0].find('span',itemprop="addressLocality").text
    data[6]=addressLocality

    addressRegion=address0[0].find('span',itemprop="addressRegion").text
    data[7]=addressRegion

    postalCode=address0[0].find('span',itemprop="postalCode").text
    data[8]=postalCode

    propertyfeatures0=soup.find('dl',class_="rui-property-features rui-clearfix")
    if propertyfeatures0 is not None:
        propertyfeatures1=propertyfeatures0.find_all('span',class_="rui-visuallyhidden")
        dds=propertyfeatures0.find_all('dd')
        for h in range(len(propertyfeatures1)):
            if propertyfeatures1[h].text=='Bedrooms':
                Bedrooms=dds[h].text
                data[9]=Bedrooms

            if propertyfeatures1[h].text=='Bathrooms':
                Bathrooms=dds[h].text
                data[10]=Bathrooms

            if propertyfeatures1[h].text=='Car Spaces':
                CarSpaces=dds[h].text
                data[11]=CarSpaces

    area0=soup.find('div',id="features")
    area1=area0.find_all('li')
    for g in range(len(area1)):
        if 'Size' in area1[g].text or 'approx' in area1[g].text:
            area=area1[g].text
            data[12]=area

    bodytext=soup.find('p',class_="body").text
    #print bodytext
    data[13]=bodytext
    data[14]=url1

    for s in range(len(data)):
        data[s]=data[s].encode('ascii', 'ignore')

    spamWriter = csv.writer(open('output4.csv', 'a'))
    spamWriter.writerow(data)
    print ("Extracting House No. Data",z)

data=['-']*15
spamWriter = csv.writer(open('output4.csv', 'a'))
spamWriter.writerow(data)
```
